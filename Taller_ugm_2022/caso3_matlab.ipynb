```matlab
dataFile = "N_202109_extent_v3.0.tif";
[I,cmap] = imread(dataFile);
imshow(I,cmap)
title("Arctic sea ice extent in September 2021")
```

```matlab
% Read the data
dataGbl = readtable("GlobalTempAnomaly.csv"); 
% Extract the year and anomaly columns
yearGbl = dataGbl.Year; 
anomalyGbl = dataGbl.Value; 
% Create the plot
plot(yearGbl,anomalyGbl) 
title("Annual Global Temperature Anomaly")
ylabel("Anomaly (°C)")
xlabel("Year")
```

```matlab
% Read the data
dataNA = readtable("NATempAnomaly.csv"); 
```

```matlab

% Extract the year and anomaly columns
yearNA = dataNA.Year; 
anomalyNA = dataNA.Value;
% Create the plot
plot(yearGbl,anomalyGbl)
hold on % Continue to plot on the same axes
plot(yearNA,anomalyNA) 
hold off 
title("Annual Temperature Anomaly")
ylabel("Anomaly (°C)")
xlabel("Year")
legend("Global","North America","Location","northwest")
```

```matlab
% Compute the moving mean
mmAnomalyGbl = movmean(anomalyGbl,10);
% Create the plot
plot(yearGbl,mmAnomalyGbl) 
title("Temperature Anomaly (10-year moving average)")
ylabel("Anomaly (°C)")
xlabel("Year")
```

```matlab
eox = wmsfind("eox");
```

```matlab

eox_marble = refine(eox,"blue marble");
```

```matlab
eox_marble_limits = refineLimits(eox_marble,"Latlim",[-90 90]);
```

```matlab
eox_update = wmsupdate(eox_marble_limits);
```

```matlab
figure
worldmap europe
mstruct = gcm;
latlim = mstruct.maplatlimit;
lonlim = mstruct.maplonlimit;

[A,R] = wmsread(eox_update,"Latlim",latlim,"Lonlim",lonlim);


geoshow(A,R)
title(eox_update.LayerTitle)
plabel off
mlabel off
```
