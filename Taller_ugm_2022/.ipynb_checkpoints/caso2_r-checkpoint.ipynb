Este notebook utiliza el paquete **aire.zmvm** descrito en  https://hoyodesmog.diegovalle.net/aire.zmvm/index.html. El paquete contiene funcciones para descargar datos de las estaciones de medicin de contaminantes viento y temperatura en la Zona Metropolitana del Valle de Mxico. 


La herramienta permite descargar las mediciones de SO2,CO,NOX,NO2,NO,O3,PM10,PM25,WSP,WDR,TMR y RH  desde el ao 1986


### Descarga de los valores mximos 2099-2021

```r
diarios de Ozono(O3) de todas las estaciones 
```

```r
library(dplyr) 
library(aire.zmvm)
```

```r
library(ggplot2)
library(ggseas)
library(magrittr)
```

 Para la  descarga se utiliza la funcin *get_station_data*

```r
o3 <- get_station_data(criterion = "MAXIMOS", 
                       pollutant = "O3", 
                       year = 2009:2021)
```

mximos y minimos diarios o promedios horarios. Las mediciones partes por billon (ppb)



```r
knitr::kable(head(o3))
knitr::kable(tail(o3))
```

Se obtienen los valores mximos de cada da

```r
o3_max <- o3 %>%
  group_by(date) %>%
  summarise(max = ifelse(all(is.na(value)),
                         NA,
                         base::max(value, na.rm = TRUE))) %>%
  na.omit()
```

```r
knitr::kable(head(o3_max))
knitr::kable(tail(o3_max))
```

 ozone threshold for declaring a 'smog alert' and the dates during which they were valid  source: http://www.aire.cdmx.gob.mx/descargas/ultima-hora/calidad-aire/pcaa/pcaa-modificaciones.pdf

```r

contingencia_levels <- data.frame(ppb = c(216, 210, 205, 199, 185, 155, 155),
  start = c(2009, 2009.4973, 2010.4973, 2011.5795,
            2012.6052,  2016.291, 2016.4986),
  end = c(2009.4973, 2010.4945, 2011.4945,
          2012.6025,    2016.2883, 2016.4959, Inf))
```

```r
contingencia_levelsdate
```

```r

max_daily_df <- tsdf(ts(o3_max$max, start = c(2009,1), frequency = 365.25))

,
```

```r
knitr::kable(head(max_daily_df))
knitr::kable(tail(max_daily_df))
```

```r
contingencia <- o3_max
,date
```

```r
contingencia$date <- max_daily_df$x
contingencia$contingencia <- case_when(
  contingencia$date > 2012.6052 & contingencia$max > 185 ~ TRUE,
  contingencia$date > 2016.291 & contingencia$max > 155 ~ TRUE,
  TRUE ~ FALSE
)
```

```r
knitr::kable(head(contingencia))
knitr::kable(tail(contingencia))
```

```r
options(repr.plot.width=15, repr.plot.height=8)
```

```r
ggplot(max_daily_df,
       aes(x = x, y = y)) +
  geom_line(colour = "grey75", alpha = .5) +,
  stat_rollapplyr(width = 30, align = "right", color = "#01C5D2") +
  geom_segment(data = contingencia_levels,
               aes(x=start, y=ppb, xend=end, yend=ppb), color="darkred",
               linetype = 2)  +
  geom_point(data=filter(contingencia, contingencia == TRUE),
             aes(x=date, y=max), color = "#999999",
             size = 2, shape = 21, fill = "firebrick3" ) +
  xlab("fecha") +
  ylab("ppb") +
  scale_x_continuous(breaks = c(2011, 2014, 2017)) +
  ggtitle("Maximum daily ozone concentration and, 30 day rolling average") +
  theme_bw()
```

Las lneas rojas en la grfica indican los valores necesarios para activar una alerta de smog de fase 1. Las lneas azules son el promedio mvil de 30 das y los puntos rojos indican cuando el valor de ozono supera el necesario para declarar una alerta de contaminacin.
