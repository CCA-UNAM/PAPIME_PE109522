# Comandos básicos de Linux

## La consola del sistema en GNU/Linux
* La consola o terminal es un programa informático que permite al usuario interactuar con el sistema operativo mediante una ventana que espera ordenes escritas desde el teclado.

* La consola permite un mayor grado de funciones y configuración de las que suele permitir un programa determinado, por ejemplo Nautilus (gestor de archivos en GNOME ).

* Es posible que varios usuarios ejecuten varias consolas simultáneamente.

* La consola ejecuta un intérprete de comandos o **shell**, el cual “traduce” los comandos de texto que escribimos para que el sistema operativo los entienda, y a la vez traduce su respuesta.

## Los intérpretes de comandos en GNU/Linux

* El intérprete de comandos es la interfaz entre el usuario y el sistema operativo, por esta razón en inglés se le da el nombre de **shell** (caparazón).

* La función del shell es leer la línea de comandos, interpretar su significado, llevar a cabo el comando y después mostrar el resultado.

## Tipos de intérpretes de comandos en Linux

Existen básicamente 4 tipos de intérpretes de comandos:

* **Compatibles con Bourne shell:** Son los más estandarizados y prácticamente todas las versiones de UNIX/Linux incluyen uno.

* **Compatibles con la shell de C:** Tiene una sintáxis similar al lenguaje de programación C.

* **Históricos:** Basados en el primer shell de UNIX (Thompson shell).

* **No tradicionales:** No caen en ninguna de las otras categorías.




## Ejemplos de intérpretes de comandos en Linux

* Compatibles con Bourne shell: sh (Bourne shell), bash (Bourne Again Shell), ash (Almquist shell), dash (Debian Almquist shell), ksh (Korn Shell).

* Compatibles con la shell de C: csh (C shell), tcsh (TENEX C), hcs (Hamilton C shell).

* Históricos: sh (Thompson shell).

* No tradicionales: fish (Friendly interactive shell), psh (Perl shell).




## Introducción a BASH

* BASH es un intérprete de comandos de UNIX escrito por el proyecto GNU.

* Su nombre es un acrónimo de Bourne-again shell (otro shell Bourne).

* Es el sistema por defecto en la mayoría de sistemas GNU/Linux, además de Mac OSX, y puede ejecutarse en la mayoría de los sistemas operativos tipo UNIX.

* También se ha portado a Windows por el proyecto Cygwin.

## Comandos para el manejo de archivos y directorios con bash


* **cd:** abrir un directorio.
* **pwd:** mostrar  la ruta en la que se encuentra ubicado.
* **ls:** mostrar contenido de directorios.
* **cat:** mostrar contenido de archivos.
* **more:** mostrar contenido de archivos controlado por el usuario.
* **mkdir:** crear un directorio.
* **rmdir:** borrar un directorio.
* **rm:** borrar un archivo vacío.
* **mv:** mover archivos (renombrar).
* **cp:** copiar archivos.
* **diff:** conocer si existe diferencia entre archivos.
* **find:** encontrar archivos/directorios.
* **grep:** puede emplearse para encontrar algo dentro de un archivo (instacias de una cadena/palabras).
* **whoami:** mostrar mi nombre de usuario.
* **touch:** crear un archivo vacío.

### Recuerda
**cd:** abre un directorio
* Para abrir un directorio determinado emplea ***cd** nombre_directorio*
* Para abrir un directorio determinado dando la ruta completa (ruta absoluta) ***cd** ruta_al_directorio/nombre_del_directorio*
* Para ir al directorio personal del usuario emplea ***cd***
* Para subir un directorio(subir un nivel) ***cd** ..*


**ls:** muestra contenido de dorectorios
* Para listar los directorios de la ubicación actual ***ls***
* Para listar los directorios de la ubicación actual en modo lista ***ls** -l*
* Para listar los directorios de la ubicación actual en modo lista, mostrando el tamaño en formato resumido ***ls** -lh*
* Para listar el contenido de un directorio en particular ***ls** nombre_directorio* o bien empleando la ruta absoluta ***ls** ruta_directorio/nombre_directorio*

## Algunas consideraciones en Linux

1. Linux distingue entre mayúsculas y minúsculas, así como con y sin acento. No se recomienda acentuar los nombres de directorios y archivos en Linux.

2. No se recomienda utilizar espacios en los nombres de directorios y archivos en Linux.

3. Los parámetros de un comando se separan por espacios, la consola reconoce la primer palabra como el nombre del programa (comando) y cada cadena de texto entre espacios como un parámetro.

4.  Cada vez que sea necesario escribir una ruta como un parámetro, se puede escribir:

    - Indicando la ruta exacta iniciando con el directorio raíz.

    - Indicando la ruta en referencia al directorio actual.

    - Indicando el directorio en referencia a alguna de las siguientes abreviaturas:

        - para el directorio actual emplea un .

        - para el directorio superior (se pueden poner varios) emplea ../

        - para la carpeta personal ~/
        
        


## Ejemplo:

Crear un directorio llamado **rpg**, crear dentro el directorio **snes**, y dentro los archivos vacíos **chrono_trigger**, **final_fantasy_iv** y **eartbound**

- cd
- mkdir rpg
- cd rpg
- mkdir snes
- cd snes
- touch chrono_trigger
- touch final_fantasy_iv
- touch earthbound
- cd
