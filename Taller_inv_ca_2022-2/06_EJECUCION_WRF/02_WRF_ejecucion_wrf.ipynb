## Ejecutando el programa *wrf.exe*

- Edita el archivo *namelist.input* para configurar las opciones de las parametrizaciones físicas a emplear, si así lo deseas.

- A partir de la versión 3.9, el modelo WRF introduce una opción para usar suites de física. Actualmente hay 1 paquete aprobado disponible llamado "CONUS", que requiere una especificación de una línea en el archivo **namelist.input**. Consiste en una combinación de opciones de física que han sido altamente probadas y han mostrado resultados buenos y razonables. 
 
- Para usar esta opción, simplemente configura el parámetro "physics_suite" dentro de la sección `&physics`

```
    physics_suite = 'CONUS'
    mp_physics =         8, 8
    cu_physics =         6, 6
    ra_lw_physics =      4, 4
    ra_sw_physics =      4, 4
    bl_pbl_physics =     2, 2
    sf_sfclay_physics =  2, 2
    sf_surface_physics = 2, 2 
```    

- Es posible anular cualquiera de las opciones anteriores por simplemente agregar ese parámetro en particular a la lista de los nombres. Por ejemplo, si deseas utilizar la suite CONUS pero deseas desactivar cu_physics para el dominio 3:

```
    physics_suite = 'CONUS'
    cu_physics = -1, -1, 0
``` 

- Si deseas usar la suite CONUS pero te gustaría usar una opción diferente de cu_physics y desactivar cu_physics para el dominio 3:

```
    physics_suite = 'CONUS'
    cu_physics = 2, 2, 0

```
- Para decidir, qué física emplear puedes revisar el manual en el Capítulo 5: WRF Model, la sección **Examples of namelists for various applications** en el manual de usuario del WRF: https://www2.mmm.ucar.edu/wrf/users/docs/user_guide_V3/user_guide_V3.9/users_guide_chap5.htm

- Para nuestro caso dejaremos las opciones que el *namelist.input* trae configurado 

- Finalmente, un parámetro muy importante es el *time_step* dentro de la sección `&time_control`. Este define el paso de tiempo, se recomienda definirlo mediante: 6xresolución_horizontal_en_km. En nuestro caso 6x9=54
