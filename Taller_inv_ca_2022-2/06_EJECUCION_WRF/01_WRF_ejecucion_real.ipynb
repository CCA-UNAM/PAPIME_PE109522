## Ejecutando el programa *real.exe*

<div align="center">

</div> 

- Para ejecutar *real.exe* es necesario haber ejecutado de forma satisfactoria los tres pasos de **WPS** y tener generados los archivos **met_em.d***

- Se requieren ligar todos los archivos de salida **met_em.d*** al directorio de ejecucin donde se encuentra el programa *real.exe*

- Entra al directorio donde se encuentra compilado el modelo WRF, en nuestro caso en 2022/wrf/WRF-4.3.3

- Ah dirjite a el directorio test/em_real

- La ruta completa sera:

     cd /home/cca/i.5010/tmp/2022/wrf/WRF-4.3.3/test/em_real

- Liga los archivos **met_em.d*** a esta direccin:

     ln s ../../../WPS-4.3.1/met_em.d* .
     
**Nota:** en la instruccin anterior es muy importamte el . para indicar el directorio en donde se encuentran, es decir el directorio actual.
