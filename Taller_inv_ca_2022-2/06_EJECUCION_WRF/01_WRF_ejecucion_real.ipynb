## Ejecutando el programa *real.exe*

<div align="center">

</div> 

- Para ejecutar *real.exe* es necesario haber ejecutado de forma satisfactoria los tres pasos de **WPS** y tener generados los archivos **met_em.d***

- Se requieren ligar todos los archivos de salida **met_em.d*** al directorio de ejecución donde se encuentra el programa *real.exe*

- Entra al directorio donde se encuentra compilado el modelo WRF, en nuestro caso en 2022/wrf/WRF-4.3.3

- Ahí diríjite a el directorio test/em_real

- La ruta completa sería:

     cd /home/cca/i.5010/tmp/2022/wrf/WRF-4.3.3/test/em_real

- Liga los archivos **met_em.d*** a esta dirección:

     ln –s ../../../WPS-4.3.1/met_em.d* .
     
**Nota:** en la instrucción anterior es muy importamte el . para indicar el directorio en donde se encuentran, es decir el directorio actual.

## Editando el archivo *namelist.input*

<div align="center">

</div> 

<div align="center">

</div> 

- Editar el archivo **namelist.input** para la sección `&time_control` configurando la fecha de inicio, fin y tiempo de integración, y la sección `&domain` para las dimensiones del dominio o dominios que se hayan definido previamente en el archivo **namelist.wps**

- En el archivo *README.namelist* se puede encontrar la descripción de las variables que se emplean en **namelist.input**. Se encuentra en la carpeta **WRF-4.3.3/test/em_real**

<div align="center">

</div> 

Es posible emplear history_interval = 60, que es equivalente a history_interval_s = 3600,   

- Ejecutar real.exe mediante el script ejecutar_real.pbs
