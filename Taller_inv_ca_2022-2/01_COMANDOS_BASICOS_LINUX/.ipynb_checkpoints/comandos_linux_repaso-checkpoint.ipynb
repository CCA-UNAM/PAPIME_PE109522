## La consola del sistema en GNU/Linux
* La consola o terminal es un programa informtico que permite al usuario interactuar con el sistema operativo mediante una ventana que espera ordenes escritas desde el teclado.

* La consola permite un mayor grado de funciones y configuracin de las que suele permitir un programa determinado, por ejemplo Nautilus (gestor de archivos en GNOME ).

* Es posible que varios usuarios ejecuten varias consolas simultneamente.

* La consola ejecuta un intrprete de comandos o **shell**, el cual traduce los comandos de texto que escribimos para que el sistema operativo los entienda, y a la vez traduce su respuesta.

## Tipos de intrpretes de comandos en Linux

Existen bsicamente 4 tipos de intrpretes de comandos:

* **Compatibles con Bourne shell:** Son los ms estandarizados y prcticamente todas las versiones de UNIX/Linux incluyen uno.

* **Compatibles con la shell de C:** Tiene una sintxis similar al lenguaje de programacin C.

* **Histricos:** Basados en el primer shell de UNIX (Thompson shell).

* **No tradicionales:** No caen en ninguna de las otras categoras.




## Ejemplos de intrpretes de comandos en Linux

* Compatibles con Bourne shell: sh (Bourne shell), bash (Bourne Again Shell), ash (Almquist shell), dash (Debian Almquist shell), ksh (Korn Shell).

* Compatibles con la shell de C: csh (C shell), tcsh (TENEX C), hcs (Hamilton C shell).

* Histricos: sh (Thompson shell).

* No tradicionales: fish (Friendly interactive shell), psh (Perl shell).




## Introduccin a BASH

* BASH es un intrprete de comandos de UNIX escrito por el proyecto GNU.

* Su nombre es un acrnimo de Bourne-again shell (otro shell Bourne).

* Es el sistema por defecto en la mayora de sistemas GNU/Linux, adems de Mac OSX, y puede ejecutarse en la mayora de los sistemas operativos tipo UNIX.

* Tambin se ha portado a Windows por el proyecto Cygwin.

## Comandos para el manejo de archivos y directorios con bash


* **cd:** abrir un directorio.
* **pwd:** mostrar  la ruta en la que se encuentra ubicado.
* **ls:** mostrar contenido de directorios.
* **cat:** mostrar contenido de archivos.
* **more:** mostrar contenido de archivos controlado por el usuario.
* **mkdir:** crear un directorio.
* **rmdir:** borrar un directorio.
* **rm:** borrar un archivo vaco.
* **mv:** mover archivos (renombrar).
* **cp:** copiar archivos.
* **diff:** conocer si existe diferencia entre archivos.
* **find:** encontrar archivos/directorios.
* **grep:** puede emplearse para encontrar algo dentro de un archivo (instacias de una cadena/palabras).
* **whoami:** mostrar mi nombre de usuario.
* **touch:** crear un archivo vaco.

### Recuerda
**cd:** abre un directorio
* Para abrir un directorio determinado emplea ***cd** nombre_directorio*
* Para abrir un directorio determinado dando la ruta completa (ruta absoluta) ***cd** ruta_al_directorio/nombre_del_directorio*
* Para ir al directorio personal del usuario emplea ***cd***
* Para subir un directorio(subir un nivel) ***cd** ..*


**ls:** muestra contenido de dorectorios
* Para listar los directorios de la ubicacin actual ***ls***
* Para listar los directorios de la ubicacin actual en modo lista ***ls** -l*
* Para listar los directorios de la ubicacin actual en modo lista, mostrando el tamao en formato resumido ***ls** -lh*
* Para listar el contenido de un directorio en particular ***ls** nombre_directorio* o bien empleando la ruta absoluta ***ls** ruta_directorio/nombre_directorio*

## Algunas consideraciones en Linux

1. Linux distingue entre maysculas y minsculas, as como con y sin acento. No se recomienda acentuar los nombres de directorios y archivos en Linux.

2. No se recomienda utilizar espacios en los nombres de directorios y archivos en Linux.

3. Los parmetros de un comando se separan por espacios, la consola reconoce la primer palabra como el nombre del programa (comando) y cada cadena de texto entre espacios como un parmetro.

4.  Cada vez que sea necesario escribir una ruta como un parmetro, se puede escribir:

    - Indicando la ruta exacta iniciando con el directorio raz.

    - Indicando la ruta en referencia al directorio actual.

    - Indicando el directorio en referencia a alguna de las siguientes abreviaturas:

        - para el directorio actual emplea un .

        - para el directorio superior (se pueden poner varios) emplea ../

        - para la carpeta personal ~/
