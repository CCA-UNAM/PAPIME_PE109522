## Modificando el archivo de configuracin **namelist.wps** para ejecutar *metgrid.exe*

* Este programa interpola horizontalmente los datos meteorolgicos extrados por *ungrib.exe* a los dominios de la simulacin definidos por *geogrid.exe*. 
* Los archivos de salida de *metgrid.exe* se nombran met_em.d0n.AAAA-MMDD_HH:mm:ss.nc, donde n es el nmero que identifica el dominio de simulacin, mm son los minutos y ss los segundos. AAAA, MM y DD son definidos como en la salida de *ungrib.exe*. 
* Modifica el archivo *namelist.wps* para leer los archivos generados por *ungrib.exe*.
    - **fg_name**: Prefijo(s) que se emplearon cuando se crearon los archivos intermedios por *ungrib.exe* 
    - **io_form_metgrid**: Define el formato en el que estarn los archivos de salida de metgrid (archivos met_em*). 2 para archivos netCDF
