### Programa para leer el archivos netcdf *geo_em.d01.nc* y graficar los dominios configurados

Carga el mapa base de la división política de México

```matlab
clear;
load 'division_mexico.dat'
```

Cargar el archivo netcdf y leer variables 

```matlab
file_nc = '../../2022/wrf/WPS-4.3.1/geo_em.d01.nc';
lat1 = ncread(file_nc,'CLAT');
long1 = ncread(file_nc,'CLONG');
```

Se construye la gráfica

```matlab
figure('Position', [1 1 662 477])
hold on
plot(long1(1,:),lat1(1,:),'Color','r','LineWidth',3)
plot(long1(:,1),lat1(:,1),'Color','r','LineWidth',3)
plot(long1(end,:),lat1(end,:),'Color','r','LineWidth',3)
plot(long1(:,end),lat1(:,end),'Color','r','LineWidth',3)
grid on

plot(division_mexico(:,1),division_mexico(:,2),'k','LineWidth',1.5)
xlim([-119.6 -84.6]);
ylim([11.06 34.85]);
box on
set(gca,'FontSize',18,'FontWeight', 'bold');
ax = gca;
ax.LineWidth = 2.0
ylabel('Latitude', 'FontSize',18, 'FontWeight', 'bold');
xlabel('Longitude', 'FontSize',18, 'FontWeight', 'bold');
text(-115,15,'Domain 1','Color','r','FontSize',20,'FontWeight','bold')

set(gcf,'PaperPositionMode','auto')
```
