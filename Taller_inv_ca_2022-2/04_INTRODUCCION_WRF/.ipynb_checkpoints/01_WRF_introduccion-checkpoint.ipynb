# Introduccin al modelo Weather Research and Forecasting (WRF)


1. Los modelos numricos atmosfricos son cada vez ms complejos y completos, y proporcionan estimaciones cada vez ms apegadas a la realidad. 

2. El modelo numrico *Weather Research and Forecasting* (WRF)  es un sistema de simulacin atmosfrica y prediccin numrica del tiempo diseado para realizar investigacin de procesos atmosfricos, y para el desarrollo de sistemas de pronstico operacional.

3. Este modelo ha sido diseado para incorporar los ltimos avances en materia de simulacin de la atmsfera y entre sus aplicaciones se encuentran:
    * Investigacin para el pronstico del tiempo, huracanes, clima regional
    * Investigacin para el desarrollo de nuevas parametrizaciones -gracias a que el cdigo permite la fcil incorporacin de nuevas esquemas de parametrizacin fsica-
    * Investigacin en asimilacin de datos y calidad del aire
    * etc.
<p> </p>
4. El desarrollo del modelo WRF inici a finales de los aos de la dcada de los 90.

5. Diversas instituciones han contribuido y siguen contribuyendo a su desarrollo (Skamarock, 2008; Wang, 2016):
    *   *National Center for Atmospheric Research* (NCAR)
    *   *National Oceanic and Atmospheric Administration* (NOAA)
    *   *National Center for Environmental Prediction* (NCEP)
    *   *Earth System Research Laboratory* (ESRL)
    *   *Air Force Weather Agency* (AFWA)
    *   *Naval Research Laboratory* (NRL) 
    *   *Center for Analysis and Prediction Storms* (CAPS) 
    *   *Federal Aviation Administration* (FAA)
    *   y la participacin de cientficos de universidades -mucho del desarrollo de la fsica del modelo es de diferentes comunidades de investigacin alrededor del mundo-  
<p> </p>
6. Es modelo WRF es de cdigo abierto (libre de usar, estudiar, distribuir y mejorar) y est disponible gratuitamente. 

7. La *Mesoscale and Microscale Meteorology Division* de NCAR es la dependencia que actualmente mantiene y respalda gran parte del cdigo del modelo. 

8. Su ejecucin puede realizarse en diferentes plataformas (computadoras porttiles hasta supercomputadoras).

9. Est diseado para ser un cdigo de computacin masivamente paralelo
    *   Las simulaciones de WRF se llevan a cabo en clsteres de computacin de alto rendimiento (HPC) -se requieren un recurso computacional eficaz que pueda manejar simulaciones complejas y paralelas-
    *   El programa *wrf.exe* es el programa que requieren mayor poder computacional   
<p> </p>
10. El hecho de que el modelo WRF pueda instalarse y ejecutarse fcilmente en hardware comn (estaciones de trabajo, computadoras porttiles y clsteres) ha sido un factor importante en su uso en la educacin.

## Componentes del modelo WRF

Los componentes del modelo WRF se presentan en la siguiente figura. Como se observa este consiste principalmente de un sistema de preprocesamiento WPS (WRF *Preprocessing System*), un proceso opcional, de asimilacin de datos (WRF-DA), el nucleo ARW y las herramientas de post-procesamiento y visualizacin de las simulaciones generadas.  

<div align="center">

</div> 

1. **WPS (Pre-procesamiento del WRF)**: es un conjunto de programas que se emplean principalmente para realizar simulaciones de datos reales. Sus principales funciones son 
    * definir los dominios de simulacin y anidamientos 
    * interpolar datos terrestres (como cobertura y uso del suelo, tipos de suelo, topografa, orogrfa, etc.) a los dominios de simulacin 
    * leer e interpolar datos meteorolgicos (formato GRIB como presin en superficie, sst, T, u, v, etc.) de otros modelos a los dominios de simulacin definidos previamente.

GRIB (GRIdded binary): Formato de datos empleado para almacenar datos de pronsticos del tiempo, con capacidad de compresin. Creado en 1985 por la WMO (World Meteorological Organization). En 1994 la nueva versin del formato 1, y en 2003 la versin 2.

2. **WRF-DA**: componente opcional encargado de introducir observaciones (por ejemplo datos de superficie, radiosondeos, datos de radar, etc.) a los anlisis interpolados que fueron creados por el WPS.

3. **El ncleo ARW (*Advanced Research WRF*)**: es el componente del modelo donde se realiza la inicializacin de la simulacin y la integracin numrica de los dominios.  
El modelo WRF tiene dos resolvedores dimmicos o ncleos para el cmputo de las ecuaciones que gobiernan la atmsfera: 
    *   WRF-ARW (*Advanced Research WRF*) desarrollado principalmente por NCAR
    *   WRF-NMM (*Nonhydrostatic Mesoscale Model*) desarrollado en NCEP
<p></p>
4. **Herramientas de post-procesamiento y visualizacin**: Componentes independientes del modelo como NCAR Command Language (NCL), Read/Interpolate/Plot (RIP4), Grid Analysis and Display System (GrADS), Vis5D, Visualization and Analysis Platform for Ocean, Atmosphere, and Solar Researchers (VAPOR), General Equilibrium Modelling Software (GEMPAK), MATrix LABoratory (MATLAB) y Python.
