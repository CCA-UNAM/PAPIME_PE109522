## El ncleo ARW
* Emplea un archivo de configuracin llamado *namelist.input*.
* El siguiente paso al WPS es el ARW, el cual es un sistema de modelado que se encarga de resolver la dinmica atmosfrica para simulaciones idealizadas y reales. 
* Los mdulos que lo componen son el real.exe y el wrf.exe. 
* Este modulo comparte un archivo de configuracin nombrado *namelist.input*, en l se configura gran parte de la informacin que se encuentra en el *namelist.wps*, pero tambin se configuran las opciones de la fsica del nucleo ARW: microfsica, cmulus, y fsica de superficie, capa lmite planetaria y la fsica de radiacin atmosfrica (longitud de onda corta y onda larga). 
* Los archivos generados por el programa *metgrid.exe* son entrada del programa *real.exe*. En este programa se interpolan verticalmente los datos a los niveles definidos dentro de archivo de configuracin *namelist.input*. 
* Finalmente, el pronstico del tiempo es generado mediante integracin numrica en el programa *wrf.exe*.
