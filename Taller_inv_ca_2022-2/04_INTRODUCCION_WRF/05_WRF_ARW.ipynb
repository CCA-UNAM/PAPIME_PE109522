## El núcleo ARW
* Emplea un archivo de configuración llamado *namelist.input*.
* El siguiente paso al WPS es el ARW, el cual es un sistema de modelado que se encarga de resolver la dinámica atmosférica para simulaciones idealizadas y reales. 
* Los módulos que lo componen son el real.exe y el wrf.exe. 
* Este modulo comparte un archivo de configuración nombrado *namelist.input*, en él se configura gran parte de la información que se encuentra en el *namelist.wps*, pero también se configuran las opciones de la física del nucleo ARW: microfísica, cúmulus, y física de superficie, capa límite planetaria y la física de radiación atmosférica (longitud de onda corta y onda larga). 
* Los archivos generados por el programa *metgrid.exe* son entrada del programa *real.exe*. En este programa se interpolan verticalmente los datos a los niveles definidos dentro de archivo de configuración *namelist.input*. 
* Finalmente, el pronóstico del tiempo es generado mediante integración numérica en el programa *wrf.exe*.
