## El archivo de configuracin namelis.wps

* Es un archivo de configuracin comn para los diferentes programas del WPS.
* Define parmetros relacionados con la definicin de los dominios, proyeccin, periodo de simulacin y formato de escritura/lectura de datos.
* Un namelist contiene una lista de opciones que el cdigo lee durante su ejecucin. Usar una namelist permite en tiempo de ejecucin cambiar la configuracin sin recompilar el cdigo fuente nuevamente.
* Contiene una zona definida para cada programa del WPS y a la vez una zona compartida. 
* Zona compartida &share con las variables usadas por ms de un programa del WPS. Por ejemplo wrf_core especifica si el WPS produce datos para el ncleo ARW o el NMM, la cual es informacin necesaria para *geogrid.exe y *metgrid.exe*.
