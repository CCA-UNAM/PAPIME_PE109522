## El archivo de configuración namelis.wps

* Es un archivo de configuración común para los diferentes programas del WPS.
* Define parámetros relacionados con la definición de los dominios, proyección, periodo de simulación y formato de escritura/lectura de datos.
* Un namelist contiene una lista de opciones que el código lee durante su ejecución. Usar una namelist permite en tiempo de ejecución cambiar la configuración sin recompilar el código fuente nuevamente.
* Contiene una zona definida para cada programa del WPS y a la vez una zona compartida. 
* Zona compartida &share con las variables usadas por más de un programa del WPS. Por ejemplo wrf_core especifica si el WPS produce datos para el núcleo ARW o el NMM, la cual es información necesaria para *geogrid.exe y *metgrid.exe*.
