#### METGRID.EXE
* Este programa interpola horizontalmente los datos meteorolgicos extrados por *ungrib.exe* a los dominios de la simulacin definidos por *geogrid.exe* (Duda, 2012). 
* El rango de las fechas que sern interpoladas por *metgrid.exe* se obtienen de la seccin share del archivo namelist.wps. 
* Los archivos de salida de *metgrid.exe* se nombran met_em.d0n.AAAA-MMDD_HH:mm:ss.nc, donde n es el nmero que identifica el dominio de simulacin, mm son los minutos y ss los segundos. AAAA, MM y DD son definidos como en la salida de *ungrib.exe*. 
* La salida de *metgrid.exe* se utiliza como entrada para el modelo WRF-ARW.
