#### UNGRIB.EXE
* La funcin principal de este programa es desempaquetar los datos meteorolgicos GRIB (GRIB1 y GRIB2) y empaquetarlos en un formato de archivo intermedio que el programa *metgrid.exe* puede leer. 
* *Ungrib.exe* no depende de *geogrid.exe* y no es dependiente de ningn dominio de simulacin definido. 
* Este se controla mediante las secciones "share" y "ungrib" del archivo namelist.wps (WRF/ARW, 2019). 
* El formato GRIB es un formato de archivo estndar de la Organizacin Meteorolgica Mundial (WMO, por sus siglas en ingls) para almacenar campos distribuidos regularmente en formato de malla. 
    * Los campos en un archivo se identifican slo por nmeros de cdigo, los cuales deben referenciarse en una tabla externa para determinar el campo correspondiente (Duda, 2012); es decir, *ungrib.exe* hace uso de tablas de Variables (Vtables dentro del WRF) para especificar qu campos descomprimir de los archivos GRIB. 
        * Los archivos Vtables enumeran los campos y sus cdigos GRIB que deben ser desempaquetados de los archivos GRIB (WRF/ARW, 2019). 
        * Tpicamente los archivos GRIB tienen muchos formatos diferentes, por lo que en el modelo WRF es posible encontrar la Vtable para los tipos de datos ms usados. 
        * Dentro de las Vtables que se encuentran definidas en el modelo se encuentran para datos GFS, CFSR NCEP, NARR, ERA-interim, entre otros. 
* Los archivos de salida de *ungrib.exe* son nombrados como FILE:AAAA-MM-DD_HH, donde AAAA es el ao de los datos en el archivo, MM es el mes, DD es el da y HH es la hora (Duda, 2012).

#### METGRID.EXE
* Este programa interpola horizontalmente los datos meteorolgicos extrados por *ungrib.exe* a los dominios de la simulacin definidos por *geogrid.exe* (Duda, 2012). 
* El rango de las fechas que sern interpoladas por *metgrid.exe* se obtienen de la seccin share del archivo namelist.wps. 
* Los archivos de salida de *metgrid.exe* se nombran met_em.d0n.AAAA-MMDD_HH:mm:ss.nc, donde n es el nmero que identifica el dominio de simulacin, mm son los minutos y ss los segundos. AAAA, MM y DD son definidos como en la salida de *ungrib.exe*. 
* La salida de *metgrid.exe* se utiliza como entrada para el modelo WRF-ARW.
