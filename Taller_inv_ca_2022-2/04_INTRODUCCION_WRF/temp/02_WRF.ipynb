## Componentes del modelo WRF

Los componentes del modelo WRF se presentan en la siguiente figura. Como se observa este consiste principalmente de un sistema de preprocesamiento WPS (WRF *Preprocessing System*), un proceso opcional, de asimilacin de datos (WRF-DA), el nucleo ARW y las herramientas de post-procesamiento y visualizacin de las simulaciones generadas.

</div> 

1. **WPS (Pre-procesamiento del WRF)**: es un conjunto de programas que se emplean principalmente para realizar simulaciones de datos reales. Sus principales funciones son 
    * definir los dominios de simulacin y anidamientos 
    * interpolar datos terrestres (como cobertura y uso del suelo, tipos de suelo, topografa, orogrfa, etc.) a los dominios de simulacin 
    * leer e interpolar datos meteorolgicos (formato GRIB como presin en superficie, sst, T, u, v, etc.) de otros modelos a los dominios de simulacin definidos previamente.

GRIB (GRIdded binary): Formato de datos empleado para almacenar datos de pronsticos del tiempo, con capacidad de compresin. Creado en 1985 por la WMO (World Meteorological Organization). En 1994 la nueva versin del formato 1, y en 2003 la versin 2.

2. **WRF-DA**: componente opcional encargado de introducir observaciones (por ejemplo datos de superficie, radiosondeos, datos de radar, etc.) a los anlisis interpolados que fueron creados por el WPS.

3. **El ncleo ARW (*Advanced Research WRF*)**: es el componente del modelo donde se realiza la inicializacin de la simulacin y la integracin numrica de los dominios.  
El modelo WRF tiene dos resolvedores dimmicos o ncleos para el cmputo de las ecuaciones que gobiernan la atmsfera: 
    *   WRF-ARW (*Advanced Research WRF*) desarrollado principalmente por NCAR
    *   WRF-NMM (*Nonhydrostatic Mesoscale Model*) desarrollado en NCEP
<p></p>
4. **Herramientas de post-procesamiento y visualizacin**: Componentes independientes del modelo como NCAR Command Language (NCL), Read/Interpolate/Plot (RIP4), Grid Analysis and Display System (GrADS), Vis5D, Visualization and Analysis Platform for Ocean, Atmosphere, and Solar Researchers (VAPOR), General Equilibrium Modelling Software (GEMPAK), MATrix LABoratory (MATLAB) y Python.
