# Editor de texto VIM

### Qu es VIM?

* Es el editor de texto que existe para **Terminal** en Linux/GNU.
* **Vim** (del ingls Vi IMrproved) es una versin mejorada del editor de texto en terminal *vi*, presente en todos los sistemas UNIX.
* Se pude encontrar en prcticamente todas las versiones de UNIX y GNU/Linux (tambin hay para OS X y Windows).
* Esta diseado para manipularse nicamente con el teclado.
* Esta hecho por programadores para programadores.
* Una vez superada la curva de aprendizaje, la productividad en este editor es superior a la productividad en un editor convencional.
* Con unas pocas combinaciones de teclas, es posible copiar, formatear y ordenar texto de muchas formas.
* Las combinaciones de instrucciones permite hacer ediciones de texto complejas.
* Est especficamente diseado para reducir el nmero de pulsaciones necesarias para editar un documento, tomando incluso en cuenta la posicin de las letras respecto de la mano.

### Modos de VIM

La principal diferencia de vim con respecto a otros editores es que trabaja con base a modos. Cada modo permite realizar un conjunto de operaciones determinadas:

Los 3 principales modos de vim son los siguientes:
* **Modo comandos**: es el modo en el que comienza vim, tambin se conoce como modo normal. En este modo se emplean combinaciones de teclas para realizar determinadas acciones de formato y edicin.
* **Modo insercin:** cuando vim esta en este modo, el texto se puede editar como en cualquier otro editor.
* **Modo lnea de rdenes**: en este modo se pueden introducir rdenes complejas, como por ejemplo buscar y reemplazar texto.

La tecla **esc** ser nuestro principal aliado para cambiar entre modos de vim.
Las teclas utilizadas para cada modo son las siguientes:
* **Modo comandos**: esc esc
* **Modo insercin**: esc i
* **Modo lnea de rdenes**: esc :

Para ejecutar vim abre una **Terminal** y escribe **vim** seguido de enter:

Algunas rdenes de vim (modo lnea de rdenes esc :)
* **:w** guardar los cambios
* **:w nombre_archivo** guardar en un nuevo archivo
* **:w!** forzar el guardado
* **:wq** guardar y salir de vim
* **:wq nombre_archivo** guardar en un nuevo archivo y salir
* **:wq!** forzar el guardado y la salida de vim
* **:q** salir de vim si no se han hecho cambios
* **:q!** forzar la salida de vim. Los cambios no guardados se pierden

Algunos movimientos en modo comandos (esc esc)
* **1G** (uno G mayscula): Ir al inicio del fichero
* **G** (G mayscula): Ir al final del fichero
* **0** (cero): Ir al inicio de la lnea
* **_$_** (signo de pesos): Ir al final de la lnea
* **Ctrl+f** (control ms f): avanzar una pgina
* **Ctrl+b** (control ms b): retroceder una pgina

### Cortar, copiar y pegar en Vim (modo comandos esc esc)

1. ***Lneas de texto***: para copiar/cortar lneas de texto colocarnos en la primera lnea que tiene el texto que nos interesa y pulsar:
* el nmero de lneas a copiar/cortar
* seguido de **yy** para copiar o **dd** para cortar 

Ejemplo: 5yy o 4dd  

Para pegar las lneas de texto situarnos donde queremos colocar el texto y pulsar:
* **p** (minscula) para pegar debajo de la lnea actual
* **P** (mayscula) para pegar encima de la lnea actual

2. ***Palabras***: para copiar/cortar palabra(s) colocarnos en la primera palabra que nos interesa y pulsar:
* **y**n**w** para copiar, en donde ***n*** es el nmero de palabras que se desean copiar
* **d**n**w** para cortar, en donde ***n*** es el nmero de palabras que se desean cortar

Para pegar palabras situarnos donde queremos colocar las palabras:
* **p** (minscula) para pegar la palabra delante del cursor
* **P** (mayscula) para pegar la palabra detrs del cursor
