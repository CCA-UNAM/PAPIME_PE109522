# Editor de texto VIM

### ¿Qué es VIM?

* Es el editor de texto que existe para **Terminal** en Linux/GNU.
* **Vim** (del inglés Vi IMrproved) es una versión mejorada del editor de texto en terminal *vi*, presente en todos los sistemas UNIX.
* Se pude encontrar en prácticamente todas las versiones de UNIX y GNU/Linux (también hay para OS X y Windows).
* Esta diseñado para manipularse únicamente con el teclado.
* Esta hecho por programadores para programadores.
* Una vez superada la curva de aprendizaje, la productividad en este editor es superior a la productividad en un editor convencional.
* Con unas pocas combinaciones de teclas, es posible copiar, formatear y ordenar texto de muchas formas.
* Las combinaciones de instrucciones permite hacer ediciones de texto complejas.
* Está específicamente diseñado para reducir el número de pulsaciones necesarias para editar un documento, tomando incluso en cuenta la posición de las letras respecto de la mano.

### Modos de VIM

La principal diferencia de vim con respecto a otros editores es que trabaja con base a modos. Cada modo permite realizar un conjunto de operaciones determinadas:

Los 3 principales modos de vim son los siguientes:
* **Modo comandos**: es el modo en el que comienza vim, también se conoce como modo normal. En este modo se emplean combinaciones de teclas para realizar determinadas acciones de formato y edición.
* **Modo inserción:** cuando vim esta en este modo, el texto se puede editar como en cualquier otro editor.
* **Modo línea de órdenes**: en este modo se pueden introducir órdenes complejas, como por ejemplo buscar y reemplazar texto.

La tecla **“esc”** será nuestro principal aliado para cambiar entre modos de vim.
Las teclas utilizadas para cada modo son las siguientes:
* **Modo comandos**: esc esc
* **Modo inserción**: esc i
* **Modo línea de órdenes**: esc :

Para ejecutar vim abre una **Terminal** y escribe **vim** seguido de enter:

Algunas órdenes de vim (modo línea de órdenes esc :)
* **:w** guardar los cambios
* **:w nombre_archivo** guardar en un nuevo archivo
* **:w!** forzar el guardado
* **:wq** guardar y salir de vim
* **:wq nombre_archivo** guardar en un nuevo archivo y salir
* **:wq!** forzar el guardado y la salida de vim
* **:q** salir de vim si no se han hecho cambios
* **:q!** forzar la salida de vim. Los cambios no guardados se pierden

Algunos movimientos en modo comandos (esc esc)
* **1G** (uno G mayúscula): Ir al inicio del fichero
* **G** (G mayúscula): Ir al final del fichero
* **0** (cero): Ir al inicio de la línea
* **_$_** (signo de pesos): Ir al final de la línea
* **Ctrl+f** (control más f): avanzar una página
* **Ctrl+b** (control más b): retroceder una página

### Cortar, copiar y pegar en Vim (modo comandos esc esc)

1. ***Líneas de texto***: para copiar/cortar líneas de texto colocarnos en la primera línea que tiene el texto que nos interesa y pulsar:
* el número de líneas a copiar/cortar
* seguido de **yy** para copiar o **dd** para cortar 

Ejemplo: 5yy o 4dd  

Para pegar las líneas de texto situarnos donde queremos colocar el texto y pulsar:
* **p** (minúscula) para pegar debajo de la línea actual
* **P** (mayúscula) para pegar encima de la línea actual

2. ***Palabras***: para copiar/cortar palabra(s) colocarnos en la primera palabra que nos interesa y pulsar:
* **y**n**w** para copiar, en donde ***n*** es el número de palabras que se desean copiar
* **d**n**w** para cortar, en donde ***n*** es el número de palabras que se desean cortar

Para pegar palabras situarnos donde queremos colocar las palabras:
* **p** (minúscula) para pegar la palabra delante del cursor
* **P** (mayúscula) para pegar la palabra detrás del cursor
